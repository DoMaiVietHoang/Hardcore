```python
def cross_entropy_loss(prob, labels):
    # Apply softmax to the logits
    probabilities = torch.softmax(prob, dim=1)
    
    # Compute the negative log probabilities for the true labels
    negative_log_probabilities = -torch.log(probabilities.gather(1, labels.unsqueeze(1)))
    
    # Compute the average loss
    loss = torch.mean(negative_log_probabilities)
    
    return loss
```
